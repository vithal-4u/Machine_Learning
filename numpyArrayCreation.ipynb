```
import numpy as np
from numpy import pi

print('@@*** int array ***@@')
a = np.array([2, 3, 4])
print(a)
print(a.dtype)

print('*** float array ***')
b = np.array([2.1, 3.2, 4.0])
print(b)
print(b.dtype)

print('*** 2D array ***')
# 2D Array
x = np.array([[1.5, 2, 3], [4, 5, 6]])
print(x)
print(x.dtype)

print('*** complex ***')
c = np.array([[1, 2], [3, 4]], dtype=complex)
print(c)

print('*** zeros ***')
print(np.zeros((3, 4)))

print('*** ones ***')
print(np.ones((2, 3, 4), dtype=np.int16))

print('*** eye ***')
print(np.eye(3))

print('*** empty ***')
print(np.empty((2, 3)))

print('*** arange ***')
d = np.arange(0, 2.5, 0.3)
print(d)

print('*** reshape ***')
x = d.reshape(3, 3)
print(x)

print('@@@*** lin space ***')
e = np.linspace(0, 2, 9)
print(e)
print('')

f = np.linspace(0, 2*pi, 90)
print(f)

print('@@@*** random number ***')
h = np.random.random((3, 4))
print(h)
print('')

k = np.random.randint(1, 25, size=(3, 4))
print(k)
print('')

l = np.random.randint(1, 25, (3, 4))
print(l)
```
